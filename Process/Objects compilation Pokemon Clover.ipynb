# Functions

```python
def process(data, cat="None", location_index=-1, item_index=-1):
    data["List_info"] = ""
    for i in range(data.shape[0]):
        if str(data.iloc[i, 0]) == 'nan': data.iloc[i, 0] = data.iloc[i-1, 0]
        data.iloc[i, data.shape[1]-1] = str(data.iloc[i, item_index]) + ": "
        indexes = [i for i in range(data.shape[1]-1) if i != location_index and i != item_index]
        if cat == 'Trades':
            data.iloc[i, data.shape[1]-1] = f'[{data.iloc[i,0]} / '+f'Nature: {data.iloc[i,1]} / '+f'Ability: {data.iloc[i,2]} / '+f'Held item: {data.iloc[i,3]} / '+f'Dresser: {data.iloc[i,4]} / '+f'ID: {data.iloc[i,5]}\n'+f'   Give: {data.iloc[i,6]}]'
        else:
            for j in indexes:
                add = str(data.iloc[i, j])
                data.iloc[i, data.shape[1]-1] += add+", " if add != 'nan' else ""
            data.iloc[i, data.shape[1]-1] = "["+str(data.iloc[i, data.shape[1]-1])[:-2]+"]"
    data = data.groupby(data.columns[location_index])["List_info"].apply(list).reset_index(name="List_info")
    data.columns=["Location","Desc"]
    data["Cat"] = cat
    data = data.iloc[:,[0,2,1]]
    return data
```

```python
def clean(data):
    for column in data.columns:
        # 
        data[column] = data[column].apply(lambda x: str(x).replace(')',''))
        data[column] = data[column].apply(lambda x: str(x).replace(' (',' | '))
        data[column] = data[column].apply(lambda x: str(x).strip())
    return data
```

```python
fr = open('Data/trainers.txt')
add = False
s = []
level_cap = 0
with open('../Gym Leaders.txt','w') as fp:
    add = False
    for line in fr:
        if 'Leader' in line: 
            add = True
            s.append(f'------ {line[:-1]} ------\n')
        if add:
            if 'Leader' not in line: s.append(line[:-1])
            if 'Level' in line :
                level = int(line.split(":")[1].strip())
                if level >= level_cap : level_cap = level
            if '===' in line:
                s.insert(1, f"Level cap : {level_cap}\n")
                for line_ in s:
                    print(line_, file=fp)
                print("\n", file=fp)
                add = False
                level_cap = 0
                s = []
```
